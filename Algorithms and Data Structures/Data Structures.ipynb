```python
# A node contains 2 pieces information, the data stored in the Node and what the Node is pointing to (next item)

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
    # Below is not necessary for implementation but needed for exam 
    def getData(self):
        return self.data
    
    def setData(self, data):
        self.data = data
        
    def getNext(self):
        return self.next

    def setNext(self, next):
        self.next = next 
        
    def __str__(self):
        if self.getNext() != None:
            return f"Data: {self.getData()}, Next: {self.getNext().getData()}"
        else: 
            return f"Data: {self.getData()}, Next: {self.getNext()}"
```

```python
# Stack implementation using python lists 

# LIFO  
class Stack():
    def __init__(self): 
        self.stack = []
        
    def push(self, data):
        self.stack.append(data)
        
    def pop(self):
        if self.stack:
            return self.stack.pop()
    
    def peek(self):
        if self.stack:
            return self.stack[-1]
    
    def size(self): 
        return len(self.stack)
    
    def __repr__(self):
        return f"{self.stack}"
    
my_stack = Stack()
my_stack.push(3)
my_stack.peek()
my_stack.pop()
my_stack.size()
print(my_stack)
```

```python
# Stack implementation using node 

class linkedListStack():
    def __init__(self):
        self.head = None # refers to most recent node 
        self.size = 0 
    
    def push(self, data):
        # create a node for the data we are abou to insert 
        node = Node(data)
        
        # set next to whatever was most recently added 
        node.next = self.head
        
        # set most recent = node 
        self.head = node
        self.size += 1
    
    def pop(self): 
        if self.size == 0: 
            return "Stack empty"
        # Head would point to the most recent node due to implementation of push 
        # head -> node -> node -> None
        #    \     ^ pop /
        #     \_________/  point head to this node 
        popped_node = self.head 
        self.head = self.head.next
        self.size -= 1
        return popped_node 
    
    def peek(self):
        if self.size == 0: 
            return f"Stack empty" 
        return self.head.data
    
    def size(self): 
        return self.size
    
    def is_empty(self): 
        return True if self.head == None else False
    
    def __repr__(self):
        """
        Return a string representation of the stack.
        """
        stack_str = "Stack: ["
        current = self.head
        while current:
            stack_str += str(current.data)
            current = current.next
            if current:
                stack_str += ", "
        stack_str += "]"
        return stack_str

        
l = linkedListStack()
# you can write your own code to test below 
l.push(1)
l.push(2)
print(l)
```

```python
# FIFO, python list implementation  
class Queue():
    def __init__(self):
        self.queue = []
    
    def enqueue(self, data):
        self.queue.append(data)
        
    def dequeue(self):
        if self.queue:
            return self.queue.pop[0]
        
    def peek(self):
        return self.queue[0]
    
    def size(self):
        return len(self.queue)
```

```python
# Circular queue using array 

class circularQueue: 
    def __init__(self, capacity): 
        self.data = [None] * self.capacity
        self.head = 0
        self.tail = self.capacity - 1
        self.capacity = capacity
        self.size = 0 
        
    def Enqueue(self, data): 
        if self.size == self.capacity: 
            return f"Queue is full" 
        if self.tail > self.head: 
            self.data[self.tail] = data
            self.size += 1
            self.tail -= 1
    
    def Dequeue(self): 
        if self.size == 0:
            return f"Queue is empty"
        else: 
            data = self.data[self.head]
            self.data[self.head] = None
            self.head += 1
            self.size -= 1
```
